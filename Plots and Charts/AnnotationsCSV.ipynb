```python
df = pd.read_csv('../LLM Testing/Final Output/elections/annotated.csv')

df
```

Correct Economic Times and Times of India

```python
def source_correction(row):
    if row['media_name'] == 'indiatimes.com':
        if 'economictimes.indiatimes.com' in row['url']:
            row['media_name'] = "economictimes." + row['media_name']
        else:
            row['media_name'] = "timesofindia." + row['media_name']

    elif row['media_name'] == 'india.com':
        if 'zeenews.india.com' in row['url']:
            row['media_name'] = "zeenews." + row['media_name']
        else:
            pass
    return row
```

```python
df = df.apply(source_correction, axis=1)

df
```

Setup Year Column

```python
df['year'] = pd.to_datetime(df['publish_date']).dt.year

df
```

### Analysis

```python
stance_data = pd.crosstab(df['year'], df['stance'])

# Define colors
colors = {'pro': 'green', 'anti': 'red', 'neutral': 'gray'}

# Create stacked bar chart
plt.figure(figsize=(12, 6))
stance_data.plot(kind='bar', 
                stacked=True,
                color=[colors.get(col, 'blue') for col in stance_data.columns],
                edgecolor='black')

plt.title('Media Stances by Year', fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.legend(title='Stance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
outlet_stance = pd.DataFrame(df.groupby('media_name')['stance'].value_counts(normalize=True).unstack(fill_value=0))
outlet_stance['cumulative_stance'] = outlet_stance.idxmax(axis=1)

outlet_stance.loc['firstpost.com', 'cumulative_stance']
```

```python
# Sort outlets by proportion of pro articles
outlet_stance = df.groupby('media_name')['stance'].value_counts(normalize=True).unstack(fill_value=0)
outlet_stance_sorted = outlet_stance.sort_values('pro', ascending=False)

# Create stacked bar chart
plt.figure(figsize=(12, 8))
ax = outlet_stance_sorted.plot(kind='barh', 
                              stacked=True,
                              color=[colors.get(col, 'blue') for col in outlet_stance_sorted.columns],
                              figsize=(12, 8))

plt.title('Media Outlets by Stance Overall', fontsize=14, fontweight='bold')
plt.xlabel('Media Outlet')
plt.ylabel('Proportion of Articles')
plt.legend(title='Stance', bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0)
plt.show()
```

#### Data

The [Digital News Report 2025](https://reutersinstitute.politics.ox.ac.uk/digital-news-report/2025) by Reuters Institute and University of Oxford was used and the CSV file was extracted, with additional outlet data embedded in the website added manually (see outlets with NaN values for the last column)

```python
df = pd.read_csv("Reference Data/data-eC82p.csv")

df
```

#### Modification of Sources

- Removals (Non-English sources, and those unavaliable on MediaCloud)
    - Remove Dainik Bhaskar
    - Remove NewsClick and DD News, as MediaCloud only hosts Hindi NewsClick (and DD News)

```python
df = df.drop([7, 9, 14])
df = df.reset_index(drop=True)

df
```

- (Identify Source IDs for each outlet and add as a column)

```python
df['MediaCloud Source IDs'] = [ 17973, 
                                271589,
                                932549, 
                                39784, 
                                282871, 
                                20258, 
                                561011, 
                                537358, 
                                22331,
                                1095, 
                                570336, 
                                56899, 
                                41238, 
                                41643, 
                                1318042, 
                                258176, 
                                282811, 
                                278493 ]

df
```

- Additions    
    - Add OpIndia, DNA (More representatives for known right wing bias)
    - Add Telegraph

```python
df = pd.concat([df, pd.DataFrame([
    {'Brand': 'OpIndia', 'MediaCloud Source IDs': 282927}, 
    {'Brand': 'DNA', 'MediaCloud Source IDs': 65929},
    {'Brand': 'The Telegraph', 'MediaCloud Source IDs': 55376},
    ])], ignore_index=True)

df
```

- Add domain names

```python
my_api_key = 'API_KEY_HERE'
```

```python
def domain_name(source_id):
    return mediacloud.api.DirectoryApi(my_api_key).source(source_id)['name']
```

```python
df['Domains'] = df['MediaCloud Source IDs'].apply(domain_name)

df
```

```python
df.loc[3, 'Domains'] = 'timesofindia.indiatimes.com'
df.loc[12, 'Domains'] = 'economictimes.indiatimes.com'

df
```

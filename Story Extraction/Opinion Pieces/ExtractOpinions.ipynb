#### Imports

```python
import pandas as pd
```

#### Load Database

```python
event =  'budget'
```

```python
df = pd.read_csv(f'../{event}.csv')

df = df.drop(columns=['Unnamed: 0'])
# missed removing index column while making the CSVs, sorry!

df
```

#### Set Up Environment

```python
synonyms = ['opinion', 'editorial', 'column', 'op-ed', 'op ed', 'analysis']
```

```python
def check_presence(text):
    text = str(text).lower()
    return any(synonym in text for synonym in synonyms)
```

```python
df_opinions = df[df['title'].apply(check_presence) | df['url'].apply(check_presence)]

df_opinions
```

```python
df_opinions.to_csv(f'{event}_opinions.csv', index=False)
```
